```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv("data_with_attendance.csv")
data
```

```python
d = data.copy()
d
```

```python
d['Attendance'] = d['Attendance'].map({'Yes':1,'No':0})
```

```python
d['Attendance']
```

```python
d
```

```python
y = d['GPA']
x1 = d[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
