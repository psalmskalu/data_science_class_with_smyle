```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv("simple_linear_regression_with_dummy.csv")
data

data2 = pd.read_csv("rand_dummy.csv")
data2
```

```python
data2.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Dummy']]


y2 = data2['GPA']
x2 = data2[['SAT','Rand 1,2,3']]

```

```python
data.isnull().sum()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
x3 = sm.add_constant(x2)
results2 = sm.OLS(y2,x3).fit()
```

```python
results2.summary()
```
