## Introduction

THis exercise involves the use of a dataset of Years of Experience and Salary with a nonlinear relationship. The dataset will be used to predict salary based on years of experience

```python
#import relevant libraries

import pandas as pd
import numpy as np


#generate the data
np.random.seed(0)

experience = np.random.uniform(1,10,100)
experience
```

```python
salary = 25000 + 5000 * experience + 800 * (experience ** 2) + np.random.normal(0,5000,100)
salary
```

```python
#Create a DataFrame
data = pd.DataFrame({'Experience':experience, 'Salary':salary})
data.head()
```

### Polynomial Regression using StatsModels

StatsModels requires us to manually create polynomial features by raising columns to their respective powers.

```python
import statsmodels.api as sm

#prepare the polynomial features upto degree 2

data['Experience^2'] = data['Experience'] ** 2
x = data[['Experience', 'Experience^2']]
y = data['Salary']

#Add a constant (intercept) for the stasmodels

x1 = sm.add_constant(x)

#fit the polynomial regression model
model = sm.OLS(y, x1).fit()

#display model summary
model.summary()
```
